```python
import pandas as pd
```

```python
data = pd.read_csv('./data/data_train.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
fig, axes = plt.subplots(4, 4, figsize=(12, 12))
false_positive_fall_paths = data[data['label'] == 1].sample(n=16)['path'].values

for ax, file_path in zip(axes.flatten(), false_positive_fall_paths):
    try:
        img = Image.open(file_path)
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(file_path.split('\\')[-2][2:])
    except Exception as e:
        ax.set_title("Error")
        ax.axis('off')
        
plt.suptitle("Fall Samples", fontsize=16, fontweight='bold', y=0.95)
plt.tight_layout()
plt.show()
```
